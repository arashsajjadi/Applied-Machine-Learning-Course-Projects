```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_Features=pd.read_csv('hiv_global_cdf_rdkit.csv')
initial_dataset=pd.read_csv('hiv.csv')
```

```python
ones_array=[]
only_ones_inf=initial_Features
for i in range(0,41127):
    if initial_dataset['HIV_active'][i]>0:
        ones_array.append(i)
    else:
        only_ones_inf=only_ones_inf.drop(i)
print('One was separated.')
```

```python
only_ones_inf_filled_mean=only_ones_inf.fillna(only_ones_inf.mean())
```

```python
only_zeros_inf=initial_Features
for i in ones_array:
    only_zeros_inf=only_zeros_inf.drop(i)
print('done')
pd.concat([only_ones_inf_filled_mean,only_zeros_inf]).sort_index()
```

```python
secondary_Features=pd.concat([only_ones_inf_filled_mean,only_zeros_inf]).sort_index().dropna()
secondary_dataset=pd.read_csv('hiv.csv')
tmp=pd.concat([only_ones_inf_filled_mean,only_zeros_inf]).sort_index().isna().any(axis=1)
np.size(tmp)
for i in np.arange(0,np.size(tmp)):
    if tmp[i]==True:
        secondary_dataset=secondary_dataset.drop(i,axis=0)
secondary_dataset
```

```python
secondary_Features
```

```python
secondary_Features=secondary_Features.loc[:,secondary_Features.apply(pd.Series.nunique) != 1]
```

```python
from sklearn.model_selection import train_test_split
X_training_data, X_test, y_training_data, y_test =train_test_split(secondary_Features,secondary_dataset, stratify=secondary_dataset['HIV_active'],test_size=0.10)
```

```python
X_training_data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=77)
principalComponents = pca.fit_transform(X_training_data)
X_training_data_PCA = pd.DataFrame(data = principalComponents)
X_training_data_PCA
```

```python
print(np.sum(pca.explained_variance_ratio_))
```

```python
y_training_data_PCA=pd.DataFrame(data =y_training_data.to_numpy()[:,1],columns=['HIV_active'])
y_training_data_PCA
```

```python
np.arange(1,6,3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,plot_confusion_matrix
knn = KNeighborsClassifier()
param_grid={'n_neighbors':np.arange(1,100,2),'metric':['euclidean','cosine','manhattan']}
grid=GridSearchCV(knn,param_grid=param_grid,cv=10,scoring='balanced_accuracy',return_train_score=False)
grid.fit(X_training_data_PCA.to_numpy().astype('float'),np.transpose(y_training_data_PCA.to_numpy().astype('int'))[0])
print("best mean cv score= ",grid.best_score_)
print("best parameters= ",grid.best_params_)
```

```python
knn_cv_results=grid.cv_results_
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(10,5),dpi=120)
plt.plot(np.arange(0,50,1),knn_cv_results['mean_test_score'][0:50],lw=1,label='euclidean',ls='-')
plt.plot(np.arange(0,50,1),knn_cv_results['mean_test_score'][50:100],lw=1,label='cosine',ls=":")
plt.plot(np.arange(0,50,1),knn_cv_results['mean_test_score'][100:150],lw=1,label='manhattan',ls='--',c='yellow')
plt.legend()
plt.grid()
plt.show()
```

```python
X_test_pca=pd.DataFrame(data=pca.transform(X_test))
grid.score(X_test_pca,y_test['HIV_active'])
```

```python
y_pred_knn=grid.best_estimator_.predict(X_test_pca)
print(confusion_matrix(y_test['HIV_active'],np.transpose(np.matrix(y_pred_knn))))
```
